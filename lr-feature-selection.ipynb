```python
df = df.drop(['Unnamed: 0'], axis=1)
```

```python
pre_email_feature_index = 12
P_emaildomain_lastindex = 70
R_emaildomain_lastindex = 130
cols = []
```

```python
def feature_selector(X, y, cols=cols):
#     y = df['isFraud']
#     X = df.iloc[:, 1:]
    lr = LogisticRegression()
    sfs = SequentialFeatureSelector(lr)
    new_features = sfs.fit_transform(X, y)
    print(sfs.n_features_to_select_, "features selected!\n")
    cols.append(X.columns[sfs.support_])
    return cols
```
